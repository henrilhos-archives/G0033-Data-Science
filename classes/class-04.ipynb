## Gerar gráfico por municipio

```python
plt.figure(figsize=(25, 7))
df.groupby('municipio_ocorrido')['municipio'].count().plot(legend=True)
plt.show()
```

O municipio de **Joinville** é o dado alvo

```python
plt.figure(figsize=(25, 7))
dfjoin = df[df['municipio_ocorrido'] == 'JOINVILLE']
dfjoin.groupby('ano')['assunto'].count().plot(legend=True)
```

```python
cols = ['municipio_ocorrido', 'municipio']
dfjoin[cols]
series = dfjoin[cols].groupby('municipio').count()
series.plot(figsize=(18, 10), fontsize=16, lw=5)
```

```python
plt.rcParams['figure.figsize'] = (35, 25)
cols = ['municipio_ocorrido', 'assunto']
plt.xlabel = "Joinville"
dfjoin[cols].groupby('assunto').count().plot.bar(fontsize=12, edgecolor='r').set(
    title = 'Demandas Por Assuntos',
    xlabel = 'Assuntos',
    ylabel = 'Quantidade'
);
plt.show()
```
